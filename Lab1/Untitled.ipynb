```python
from matplotlib import pyplot as plt   
from matplotlib import style   
import numpy as np
style.use('ggplot')
```

```python
d = [0.267, 0.523, 0.670, 0.709, 0.806]
d_o = 0.782
c = [10/1500, 20/1500, 30/1500, 40/1500, 50/1500]
```

```python
p = np.polyfit(c, d, 1)
yp = np.polyval(p, c)
```

```python
xp_o = (d_o - p[1])/p[0]
```

```python
plt.scatter(c, d)
plt.scatter(xp_o, d_o)
plt.plot(c, yp, 'r', label='Approx', color="g")

plt.title('')
plt.ylabel('D')
plt.xlabel('C')

plt.show()
```
